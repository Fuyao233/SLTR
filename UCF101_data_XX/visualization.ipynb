```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
```

```python
group_list = np.arange(10)
for i in group_list:
    print(f'Loading data group{i}!')
    data = loadmat(f'group{i}.mat')['mid'].toarray()
    print(f'Finish!')
    data = np.transpose(data).reshape(-1)
    plt.hist(data, 100, weights=np.zeros_like(data) + 1 / len(data))
    plt.xlabel('value')
    plt.ylabel('Frequency')
    plt.title(f'group{i} mean:{np.mean(data)}')
    plt.savefig(f'group{i}.png')
    plt.figure()
```
